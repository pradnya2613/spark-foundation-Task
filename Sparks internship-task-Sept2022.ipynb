**Aim :** Predict the percentage of a student based on the number of study hours

**Imporing the libraries**

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

**Importing Dataset**

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")

data.head(10)
```

**Exploring the data**

```python
#Check the shape of the data
data.shape
```

```python
#Check the Information of the data
data.info()
```

```python
#Check the description of the data
data.describe()
```

**Visualization of the data**

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage', color="green", size=22)  
plt.xlabel('Hours Studied',size=15)  
plt.ylabel('Percentage Score', size=15)  
plt.show()
```

We can clearly see that there is a positive linear relation between the number of hours studid and percentage of student.

**Preparing the data**

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values  
print("Data is successfully divided into attributes and labels")
print("X is independent variable and Y is dependant variable")
```

**Splitting the data into Training and Testing Set**

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
X_test.shape
```

**Training the model**

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training completed!")
```

```python
#Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y, s=100)
plt.plot(X, line, color="red");
plt.show()
```

**Making Predictions**

```python
# Predicting the scores 
y_pred = regressor.predict(X_test)
#predicted output
y_pred
```

```python
#Actual Output
y_test
```

```python
hour = np.array([9.25])
score = regressor.predict(hour.reshape(-1, 1))
print("No. of study hours: {0} and Score : {1}". format(hour, score))

# own_pred = regressor.predict(hours)
# print("No of Hours = {}".format(hours))
# print("Predicted Score = {}".format(own_pred[0]))
```

```python
m = regressor.coef_
c = regressor.intercept_
print("Coefficiet : {0} and Intercept : {1}".format(m, c))
```

```python
# y = m*x + c
score = m*(9.25) + c
score
```

**Evaluating the model**

```python
from sklearn import metrics
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
```

```python
#Identify the accuracy
regressor.score(X, y)
```
